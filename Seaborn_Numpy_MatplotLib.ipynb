### importing the libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Create a figure with subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 8))

# Plot the first distribution in the top-left subplot
sns.distplot(np.random.normal(loc=1, scale=2, size=1000), hist=True, ax=axs[0])
axs[0].set_title('Normal Distribution')

# Plot the second distribution in the top-right subplot
sns.distplot(np.random.binomial(n=10, p=0.5, size=1000), hist=True, kde=True, ax=axs[1], color='red')
axs[1].set_title('Binomial Distribution')

# Show the plots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of squared values using np.arange and a generator expression
y = np.array([i**2 for i in range(-10, 10)])
x = np.array([i for i in range(-10, 10)])
print(x, y)
# Plot the array
plt.plot(x, y, 'o:r', mec='g', mfc='hotpink')

# Set title with custom font properties
font_title = {'family': 'serif', 'color': 'yellow', 'size': 20}
plt.title("X Square Graph", loc='center', fontdict=font_title)

# Label the axes
plt.xlabel('X-Values')
plt.ylabel('Squared-Values')

# Add a grid
plt.grid()

# Display the plot
plt.show()
```
