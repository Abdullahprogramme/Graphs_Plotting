### A simple graph

```python
import matplotlib.pyplot as plt
import numpy as np
import random

x = np.array(random.sample(range(1,9), 4))
y = np.array(random.sample(range(1,9), 4))

plt.plot(x, y, 'o-.c')
plt.grid()
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.show()
```

## bar graph

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([1, 2, 3, 4])
colors = np.array([10, 20, 50, 84])

# Use the 'color' parameter instead of 'c'
bars = plt.bar(x, y, color=plt.cm.viridis(colors / max(colors)))
plt.grid(axis='y')
plt.xlabel("Numbers", loc="right")
plt.ylabel("Letters", loc="bottom")
plt.legend(bars, ['Category A', 'Category B', 'Category C', 'Category D'], loc='upper left')
plt.show()
```

## histogram

```python
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```

## pie chart

```python
y = np.array([12, 34, 90, 9])
print(y[1]/sum(y))
my_labels = ["A", "B", "C", "D"]
explosion = [0.05, 0.05, 0.05, 0.05]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
plt.pie(y, labels=my_labels, startangle= 90, explode= explosion, shadow=True, colors=mycolors, autopct='%1.1f%%', textprops=dict(color='red'))
plt.legend(loc="upper left")
plt.show()
```
