```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading the csv file
df = pd.read_csv('C:\Personal Files\OneDrive - Habib University\Python\Jupyter NoteBooks\waterQuality1.csv')

# droppping the cells which have no value
df.dropna(subset=['ammonia'], inplace=True)
df = df[df['ammonia'] != '#NUM!']
df = df[df['nitrites'] != '#NUM!']

# showing the data in string format
print(df.to_string())

df['ammonia'] = pd.to_numeric(df['ammonia']) 
df['nitrites'] = pd.to_numeric(df['nitrites']) 
print(df['ammonia'].mean())
```

```python
print(df.duplicated())

plt.plot(df['ammonia'])
plt.xlabel('Index')
plt.ylabel('Ammonia Values')
plt.title('Ammonia Values Over Index')
plt.show()
```

#### making a graph of ammonia column of month each (31 days).

```python
x_values = []  # List to store x-axis values
amm_y_values = []  # List to store y-axis values
nit_y_values = []

for i in range(0, 124, 31):
    subset_df = df.iloc[i:i+31]
    amm_average = subset_df['ammonia'].mean()
    nit_average = subset_df['nitrites'].mean()

    # Append x and y values to the lists
    x_values.append(i + 15)  # Use the midpoint of the subset as the x-value
    amm_y_values.append(amm_average)
    nit_y_values.append(nit_average)

# Plot the line connecting the mean points
plt.plot(x_values, amm_y_values, marker='o', linestyle='-', label='Ammonia Average')
plt.plot(x_values, nit_y_values, marker='o', linestyle='-', label='Nitrites Average')

plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Values Over Index, first 124 days')
plt.grid()
plt.legend()
plt.show()
```

#### making a graph of all columns all together

```python
# making a list of all column headings
headings = df.columns.tolist()
print(headings)
# looping through the monthly values
for head in headings[:-1]:
    y_values = []  # List to store y-axis values
    x_values = []  # List to store x-axis values
    for i in range(0, 124, 31):
        subset_df = df.iloc[i:i+31]
        average = subset_df[head].mean()
        y_values.append(average)
        x_values.append(i + 15)  # Use the midpoint of the subset as the x-value
        
    # Plot the line connecting the mean points
    print(y_values, x_values)
    plt.plot(x_values, y_values, marker='o', linestyle='-', label=f"{head} Average")

plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Values Over Index, first 124 days')
plt.grid()
#plt.legend()
plt.show()
```
