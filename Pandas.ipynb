```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading the csv file
df = pd.read_csv('C:\Personal Files\OneDrive - Habib University\Python\Jupyter NoteBooks\waterQuality1.csv')

# droppping the cells which have no value
df.dropna(subset=['ammonia'], inplace=True)
df = df[df['ammonia'] != '#NUM!']
df = df[df['nitrites'] != '#NUM!']

# showing the data in string format
print(df.to_string())

df['ammonia'] = pd.to_numeric(df['ammonia']) 
df['nitrites'] = pd.to_numeric(df['nitrites']) 
print(df['ammonia'].mean())
```

```python
print(df.duplicated())

plt.plot(df['ammonia'])
plt.xlabel('Index')
plt.ylabel('Ammonia Values')
plt.title('Ammonia Values Over Index')
plt.show()
```

#### making a graph of ammonia column of month each (31 days).

```python
x_values = []  # List to store x-axis values
amm_y_values = []  # List to store y-axis values
nit_y_values = []

for i in range(0, 124, 31):
    subset_df = df.iloc[i:i+31]
    amm_average = subset_df['ammonia'].mean()
    nit_average = subset_df['nitrites'].mean()

    # Append x and y values to the lists
    x_values.append(i + 15)  # Use the midpoint of the subset as the x-value
    amm_y_values.append(amm_average)
    nit_y_values.append(nit_average)

# Plot the line connecting the mean points
plt.plot(x_values, amm_y_values, marker='o', linestyle='-', label='Ammonia Average')
plt.plot(x_values, nit_y_values, marker='o', linestyle='-', label='Nitrites Average')

plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Values Over Index, first 124 days')
plt.grid()
plt.legend()
plt.show()
```

#### making a graph of all columns all together

```python
# making a list of all column headings
headings = df.columns.tolist()
# looping through the monthly values
for head in headings[:-1]:
    y_values = []  # List to store y-axis values
    x_values = []  # List to store x-axis values
    for i in range(0, 217, 31):
        subset_df = df.iloc[i:i+31]
        average = subset_df[head].mean()
        if average > 2:
            y_values.append(average)
            x_values.append(i + 15)  # Use the midpoint of the subset as the x-value
        
    # Plot the line connecting the mean points
    plt.plot(x_values, y_values, marker='o', linestyle='-', label=f"{head} Average")

plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Values Over Index, first 217 days')
plt.grid()
#plt.legend()
plt.show()
```

### Apple stock data graphs

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# reading the csv file
df = pd.read_csv('C:\Personal Files\OneDrive - Habib University\Python\Jupyter NoteBooks\\apple_stock.csv')
# dropping any empty cell
df.dropna(inplace=True)

print(df.to_string())

```

```python
df['Date'] = pd.to_datetime(df['Date'])
filtered_df = df[df['Date'].dt.year >= 2023]
print(filtered_df.to_string())

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(14, 14))

# Iterate through each column and plot monthly means
for i, column in enumerate(['Open', 'High', 'Low', 'Close']):
    row_index = i // 2
    col_index = i % 2
    # Group by month and calculate the mean for each group for the current column
    monthly_means = filtered_df.groupby(filtered_df['Date'].dt.to_period("M"))[column].mean()
    # Plot the monthly means for the current column on a separate subplot with points
    monthly_means.plot(kind='line', xlabel='Month', ylabel='Mean Value', title=column, marker='o', mfc='hotpink', ax=axes[row_index, col_index], colormap='viridis')
    monthly_means.plot(kind='line', xlabel='Month', ylabel='Mean Value', title='All', marker='o', mfc='yellow', ax=axes[2, 0], grid=True)

axes[2, 1].axis('off')
axes[2, 0].legend()
plt.tight_layout()
plt.show(block=False)
```

```python
filtered_df['Volume'] = pd.to_numeric(filtered_df['Volume'])
filtered_df['Volume'] = filtered_df['Volume'] / 1000
print(filtered_df.to_string())
monthly_mean = filtered_df.groupby(filtered_df['Date'].dt.to_period('M'))['Volume'].mean()
print('?????',monthly_mean)
monthly_mean.plot(ylabel='Volume means', xlabel='Months', grid=True, title='Mean', kind='line', marker='o')
plt.show()
```
